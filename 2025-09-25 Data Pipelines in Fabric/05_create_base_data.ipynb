# 05: Create Base Table

Finally, we create a base table by joining our end results for patients and appointments together.

```python
%run 00_config
```

```python
# Read the data from the lakehouse
df = spark.sql(f"""
SELECT
    p.*,
    IFNULL(count_appointments, 0) AS count_appointments,
    latest_appointment,
    earliest_appointment,
    IFNULL(count_doctors_seen, 0) AS count_doctors_seen
FROM 
    {config['lakehouse_name']}.personal_df_clean p
LEFT JOIN
    {config['lakehouse_name']}.appointment_df_agg a ON p.patient_id = a.patient_id
""")
display(df)
```

```python
# Save the data to the lakehouse
(
df
    .write
    .mode('overwrite')
    .format('delta')
    .option('overwriteSchema', 'true')
    .save(f"{config['lakehouse_path']}/base_table")
)
```
